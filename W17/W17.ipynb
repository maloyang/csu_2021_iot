## 以B2的CO2為例子

因為資料每5分鐘一筆，所以請依次讀取資料不要超過4小時為佳，若過多的資料再請自行排版

```python
# select
resultProxy=my_db.execute("select * from data_table where sid='B2' and sensor='CO2' and dtime>='2021-12-21 12:00:00'")
data = resultProxy.fetchall()

for item in data:
  print(dict(item))
```

```python

#先整理資料
time_list = []
my_data_list = []
for item in data:
  try:
    time_list.append(item['dtime'])
    my_data_list.append(int(item['value']))
  except Exception as e:
    print('error: ', str(e))
    my_data_list.append(None)
print(len(my_data_list))
```

```python
# 先畫圖，不存檔
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(time_list, my_data_list)
plt.xlabel('time')
plt.ylabel('CO2')
plt.xticks(time_list[::4], rotation=90)
plt.grid()
```

```python
# 畫圖並存檔
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(time_list, my_data_list)
plt.xlabel('time')
plt.ylabel('CO2')
plt.xticks(time_list[::4], rotation=90)
plt.grid()

fn = 'image1.png'
plt.savefig(fn)
plt.close()
```
